```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
#Paranetros de la distribucion binomial:
n = 10  #Numero de ensayos.
p = 0.25  #probabilidad de exito.

#Creamos un rango de valores posibles:
k = np.arange(0, n+1)

#Calculamos las probabilidades para cada valor:
probas = binom.pmf(k, n, p)

media = n*p
desviacion = np.sqrt(n*p*(1-p))
```

```python
plt.bar(k, probas, color = 'lightblue', edgecolor='blue')
plt.axvline(media, color = 'red', linestyle='--', label=f'Media = {media:.2f}')
plt.axvline(media + desviacion, color = 'green', linestyle='--', label=f'Desviacion = +/-({desviacion:.2f})')
plt.axvline(media - desviacion, color = 'green', linestyle='--')
plt.title(f'Distribucion Binomial (n={n}, p={p})')
plt.xlabel('Numero de exitos')
plt.ylabel('Probabilidad')
plt.xticks(k)
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.show()
```

```python
n1 = 10
p1 = 0.9

proba = binom.pmf(k, n1, p1)
mu = n1*p1
std = np.sqrt(n1*p1*(1-p1))

plt.bar(k, proba, edgecolor = 'black')
plt.axvline(mu, color = 'red', linestyle='--', label=f'Media = {mu:.2f}')
plt.axvline(mu+std, color = 'green', linestyle = '--', label=f'Desviacion = +/-({std:.2f})')
plt.axvline(mu-std, color = 'green', linestyle='--')
plt.title(f'Distribucion Binomial (n = {n1}, p={p1})')
plt.xlabel('Numero de exitos')
plt.ylabel('Probabilidad')
plt.legend()
plt.xticks(k)
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.show()
```

```python
#Aqui vamos a hacer lo mismo utlilizanfo seaborn
import seaborn as sns
n2 = 15
p2 = 0.3

k2 = np.arange(0, n2+1)
pro = binom.pmf(k2, n2, p2)

sns.barplot(x = k2, y = pro, color = 'skyblue')
plt.title('Distribucion Binomial (n = {n2}, p = {p2})')
plt.xlabel('Numero de exitos')
plt.ylabel('Probabilidad')
plt.xticks(rotation =45)
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
ns = 20
ps = [0.25, 0.5, 0.75]
ks = np.arange(0, ns+1)

plt.figure(figsize = (10,6))

for p in ps:
  pbs = binom.pmf(ks, ns, p)
  mus = ns*p
  plt.plot(ks, pbs, 'o-', label = f'p = {p}')
  plt.axvline(mus, color = 'black', linestyle = '--', label = f'Media de para p = {p}')



plt.title(f'Distribuciones Binomiales para n = {ns}')
plt.xlabel('Numero de exitos (k)')
plt.ylabel('Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
