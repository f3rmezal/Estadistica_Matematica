```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lambdaa = 1
k = np.arange(0,20)
pro = poisson.pmf(k, lambdaa)
```

```python
plt.bar(k, pro)
plt.title(f'Distribucion de Poisson (Î» = {lambdaa})')
plt.xlabel('Numero de eventos (k)')
plt.ylabel('Probabilidad')
plt.xticks(k)
plt.grid(axis = 'y', alpha = 0.5)
plt.show()
```

```python
#Utilizemos distintos valores de lambda en un mismo grafico:

lambdas = [1, 4, 10]
colors = ['b', 'g', 'r']
k1 = np.arange(0,30)

plt.figure(figsize=(10, 6))


for l, c in zip(lambdas, colors):
  plt.bar(k1, poisson.pmf(k1, l), color = c, alpha = 0.6,
          label = f'Î» = {l}', width=0.8)

plt.title('Distribuciones de Poisson para diferentes Î»')
plt.xlabel('Numero de eventos (k)')
plt.ylabel('Probabilidad P(X=k)')
plt.legend()
plt.grid(axis = 'y', alpha=0.5)
plt.xticks(k)
plt.show()
```
